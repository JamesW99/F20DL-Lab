```python
# add 4 column in weather date
cw.insert(0,'year',0)
cw.insert(1,'month',0)
cw.insert(2,'day',0)
cw.insert(3,'hour',0)
cw
```

```python
import re
lens = len(cw['DATE'])
for m in range(lens):
    matchObj = re.match(r'(0|[1-9][0-9]*)/(0|[1-9][0-9]*)/(0|[1-9][0-9]*) (0|[1-9][0-9]*):.*', cw['DATE'][m], re.M | re.I)

#     print(f'{matchObj.group()} - Month: {matchObj.group(1)}, Day: {matchObj.group(2)}, Year: {matchObj.group(3)}, Hour: {matchObj.group(4)}')
#     print("Month: ", matchObj.group(1))
#     print("Day: ", matchObj.group(2))
#     print("Year: ", matchObj.group(3))
#     print("Hour: ", matchObj.group(4))
    cw['year'][m] = matchObj.group(3)
    cw['month'][m] = matchObj.group(1)
    cw['day'][m] = matchObj.group(2)
    cw['hour'][m] = matchObj.group(4)
cw
```
