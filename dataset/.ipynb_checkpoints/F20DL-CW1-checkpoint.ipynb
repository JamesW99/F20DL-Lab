## binning

```python
ontime = 0
ok = 0
worried = 0
annoying = 0
angry = 0

for i in data['dep_delay']:
    if(i <= 0):
        ontime+=1;
    elif(i <= 10):
        ok+=1;
    elif(i <= 30):
        worried +=1;
    elif(i <= 90):
        annoying +=1;
    else:
        angry+=1;

print(ontime)
print(ok)
print(worried)
print(annoying)
print(angry)
```

```python
from sklearn.model_selection import train_test_split
data = pd.read_csv('new.csv')
X = data.drop(['arr_delay'], axis = 1)
y = pd.read_csv('new.csv', usecols=['arr_delay'])

X_train ,X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
print ('Accuracy:', accuracy_score(y_test, y_pred))
print('F measure:', f1_score(y_test, y_pred, average="micro"))
print ('Precision:', precision_score(y_test, y_pred, average="micro"))
print ('Recall:', recall_score(y_test, y_pred, average="micro"))
print ('mean squared error:', mean_squared_error(y_test, y_pred))
```

```python
print ('clasification report:\n', classification_report(y_test, y_pred))
```
